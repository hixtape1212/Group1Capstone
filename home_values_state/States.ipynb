### In this notebook I tried to find out which states had a bigger increse in real state prices 

```python
import pandas as pd
```

```python
df=pd.read_csv('State_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month (1).csv')
```

```python
# Get earinings between difference in price from 2000 to 2023 
df['difference']=df['2023-06-30']-df['2000-01-31']
```

```python
# States order by earnings
df_states=df[['RegionName','difference']].sort_values(by='difference', ascending=False)
df_states
```

```python
df_states=df_states.set_index('RegionName')
```

```python
#plt.figure(figsize=(100000, 10000000))
import matplotlib.pyplot as plt
df_states.plot(kind='bar')
plt.rcParams['figure.figsize'] =[18,10]
plt.show()
```
