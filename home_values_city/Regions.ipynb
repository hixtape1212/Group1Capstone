### In this notebook I tried to find out which regiones had a bigger increse in real state prices in each state

```python
import pandas as pd
```

```python
df=pd.read_csv('City_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
# number of regions in each state
df['State'].value_counts()
```

```python
# Get earnings between difference in price from 2000 to 2023 
df['difference']=df['2023-06-30']-df['2000-01-31']
```

```python
df_state=df[['RegionName', 'State', 'difference']]
```

```python
df_state
```

```python
# deleting NaN
df_state=df_state.dropna()
```

```python
df_state.isna().sum()
```

```python
type(df_state)
```

```python
df_state.sort_values(by='State')
```

```python
# number of regions in each state
df_state['State'].value_counts()
```

```python
# check if there are more than one RegionName in each State
filt=['State','RegionName']
df_state[filt].value_counts()
```

```python
# Get specific RegionName in each State
df_state[ (df_state['State']=='OH') &  (df_state['RegionName']=='Oakwood') ]
```

```python
# Group for State and get the mean for each RegionName
mean_RegionName=df_state.groupby(['State','RegionName'])['difference'].mean()
mean_RegionName
```

```python
type(mean_RegionName)
```

```python
df_mean_RegionName=mean_RegionName.to_frame()
df_mean_RegionName
```

```python
# Filter per state, in this case Texas
df_mean_RegionName.loc['TX'].sort_values(by='difference', ascending=False)
```
